```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
#DataSet Import
housedata = pd.read_csv('Housing.csv')
housedata
```

```python
#Cleaning
# housedata.loc[0,"bedrooms"] = 4
housedata.isnull().sum()
```

```python
#Anlyzing

def analyze(data):
   if(data=="yes"):
     return 1
   else:
      return 0
housedata["mainroad"] = housedata["mainroad"].apply(analyze)
housedata
```

```python
housedata["guestroom"] = housedata["guestroom"].apply(analyze)
housedata["basement"] = housedata["basement"].apply(analyze)
housedata["hotwaterheating"] = housedata["hotwaterheating"].apply(analyze)
housedata["airconditioning"] = housedata["airconditioning"].apply(analyze)
housedata["prefarea"] = housedata["prefarea"].apply(analyze)

def furnished(data):
   if(data=="unfurnished"):
     return 1
   elif(data=="semi-furnished"):
      return 2
   else:
      return 3
housedata["furnishingstatus"] = housedata["furnishingstatus"].apply(furnished)
housedata
```

```python
housedata.info()
```

```python
X = housedata.drop("price",axis=1)

Y = housedata["price"]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
X_test.shape[0]
```

```python
MyfirstModel= LinearRegression()
```

```python
MyfirstModel.fit(X_train,Y_train)
```

```python
Y_predicted =  MyfirstModel.predict(X_test)
r2 = r2_score(Y_test, Y_predicted)
print(str(int(r2*100))+"%")

Y_predicted = [int(value) for value in Y_predicted]

df_predicted = pd.DataFrame({
    'Y_predicted': Y_predicted,
    'Y Original': Y_test
})
df_predicted
```

```python

features = {
    'area': 2000,
    'bedrooms': 5,
    'bathrooms': 4,
    'stories': 2,
    'mainroad': 1,
    'guestroom': 1,
    'basement': 0,
    'hotwaterheating': 1,
    'airconditioning': 1,
    'parking': 1,
    'prefarea': 1,
    'furnishingstatus': 3
}

coefficients = {
    'intercept': 80000,
    'area': 1000,
    'bedrooms': 20000,
    'bathrooms': 15000,
    'stories': 50000,
    'mainroad': 50000,
    'guestroom': 30000,
    'basement': -20000,
    'hotwaterheating': 25000,
    'airconditioning': 30000,
    'parking': 15000,
    'prefarea': 40000,
    'furnishingstatus': 10000
}


def manual_prediction(features, coefficients):
    prediction = coefficients['intercept']
    for feature, value in features.items():
        prediction += coefficients[feature] * value
    return prediction


predicted_price = manual_prediction(features, coefficients)


print(f"Predicted Price: {predicted_price}$")
```
